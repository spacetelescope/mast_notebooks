# Subscriptions
---
This is a short notebook, focused on processing the results of a MAST subscription email notification. By the end of this notebook, you will be able to you:
- Use astroquery to download products from a subscription notification
- Categorize subscriptions, based on the contents of the JSON blob.


## Imports

There are limited packages needed for this notebook:
- `astroquery.mast` to download products. For the latest functionality (and to run the two-line download), you'll need to install directly from the repo: `pip install git+https://github.com/astropy/astroquery.git`
- (optional) `json` to open the json blob contained in the subscriptions email

## Quick Run: Two-Line Download

If you are simply interested in getting your products as quickly as possible, this can be done in just two lines of code. All you need to do is point `download_products` at the file you received in your email:

```python
from astroquery.mast import MastMissions
MastMissions.download_products("JWST.json")
```

As with a standard astroquery download, the display will show a set of download notifications, along with a status table of all products after completion.

That's it! You're done. Your products are downloaded.

## A Little More Detail

Although it is easy to download your products, it is useful to know more about how the download proceeds and the contents of the JSON file.

### Examining the JSON Contents

Let's a look at the information in the JSON itself. We'll need one more import to accomplish this: `json` will read our file into a list of dictionaries.

```python
# we need this to properly read in the file
import json

# open our file and take a look inside
with open("JWST.json", "r") as infile:
    subscriptions = json.load(infile)
subscriptions
```

As we would expect from our download., there are three files in this list.

Although we can see the entire file here, let's briefly simplify this to understand what information is being provided to us.

```python
# get the keys of the first Python dictionary
keys = subscriptions[0].keys()
# print out the keys
print(len(keys), keys)
```

Of these eight keys, it is particularly important to understand the `event_type`, which represents the reason for the subscription notification:
- `Reprocessed`: the file was reprocessed by an updated version of a calibration pipeline
- `Public`: the file was previously under an [exclusive access period (EAP)](https://archive.stsci.edu/publishing/data-use#h2-24a4fb60-93bc-4b19-803e-e8e7b04e8823), but is now publicly available
- `New`: this file is new to the archive

In the case where an investigator waives their EAP, files will be considered both new **and** public; notifications for either category will trigger.


```python
for s in subscriptions:
    print(s['event_type'])
```

### A Note on the Download Process

Normally, when using the `MissionMast` module, you need to specify the mission to complete a download. For example, if we try to download the first file in our set (`jwst.download_file("jw05893001001_02201_00001_mirimage_o001_crf.fits`):

```python
# specify mission first
jwst = MastMissions(mission='jwst')
# now download file
jwst.download_file("jw05893001001_02201_00001_mirimage_o001_crf.fits")
```

Since one of the JSON keys is `mission`, MAST has designed the `download_products` function to read from this directly; no need to add this value manually.
