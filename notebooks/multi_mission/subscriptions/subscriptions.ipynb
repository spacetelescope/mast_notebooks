## imports

straightforward for this notebook
- `astroquery.mast` to download products
- `json` to open the json blob contained in the subscriptions email

```python
from astroquery.mast import MastMissions
import json
```

```python
MastMissions.session_info()
```

## examining the JSON
take a look at the information in the JSON returned by the email

```python
with open("JWST.json", "r") as infile:
    subscriptions = json.load(infile)
subscriptions
```

we've simplified this JSON file to contain only two examples. the most important category to understand is `event_type`, which represents the reason for the subscription notification:
- `Reprocessed`: the file was reprocessed by an updated version of the calibration pipeline
- `Public`: the file was previously under and EAP (???) but is newly publicly available (fact check needed)
- `New`: this file is new to the archive

## downloading products
since we're using the `MissionsMast` module, we first need to specify our mission of interest. in this case, we were subscribed to JWST data, so we need to set that value.

```python
# don't forget to set the mission
jwst = MastMissions(mission='jwst')
```

with this initialized, it's a single step to call the `download_products` function, using our json file as an argument.

```python
jwst.download_products("JWST.json")
```

wow yay we downloaded our data!
